```python
import matplotlib.pyplot as plt
import numpy as numpy
import time
import random as rd
def crue(NIVEAU,ETAT,L,V,t):#L=largeur du fleuve, V=vitesse d'écoulement,
#a=importance de la croissance exponentielle,m=temps de stagnation du niveau,
#t=temps de l'expérience, ajouter porosité     
    '''d={0:[1,1.85] , 1:[1,1.85] , 2:[1,1.85] , 3:[1,1.85] , 4:[1,1.85] , 5:[1,1.85] , 6:[1,1.85],
    7:[1,1.85] , 8:[1,1.85] , 9:[1,1.85] , 10:[1,1.85] , 11:[1,1.1], 12:[0,1] , 13:[3,1.9]} #{id:[altitude,coef de ruissellement]}
    '''
    F=ETAT[2]
    D=L*V
    barrage=input("Présence en amont d'un barrage? Oui=1, Non=0")
    if int(barrage)==1:
        D=1
    sécheresse=input('Contexte de sécheresse ? Oui=1, Non=0')
    if int(sécheresse)==1 and ETAT[0]==13:
        s=rd.randint(105,145)
        F=F*(s/100)
    a=rd.randint(100,150)/100
    if V==1:
        a=1
    
    m=rd.randint(int((1/6)*t),int((1/4)*t))
    msg=1
    aff=0
    x = []
    y = [] #On créé deux listes vides pour contenir les abscisses et les ordonnées
    pas = 1
    abscisse = 0 #L'abscisse de départ est 0
    i=1
    r=0 #nombre choisi au hasard pour la pluie
    p=0 #temps averse
    intensité=0 #intensité des précipitations
    while i<=m:
        if int(sécheresse)==1:
            
            NIVEAU[0]=NIVEAU[0]*numpy.exp(-0.00001+((a*D)-D)*(1/(i+50000)))#-0.00001=coeff de décroissance
            x.append(abscisse+1)
            y.append(NIVEAU[0])
            abscisse += pas
            i=i+1
            
            
            
        NIVEAU[0]=NIVEAU[0]*numpy.exp(((F*a*D-D))*(1/(i+50000))) #recherche d'une fonction satisfaisante
        x.append(abscisse+1)
        y.append(NIVEAU[0])
        abscisse += pas
        
        if (i/1000)%1==0:
            i=i+1
            #print(NIVEAU[0]) #?
        else:
            i=i+1
            #print(NIVEAU[0])
        if msg==1 and NIVEAU[0]>=ETAT[1]: #attention chiffre dépendant de l'altitude
            msg=0
            print('Inondation en cours') #autres alertes ?
            
    while i<=2*m: #revoir condition
        x.append(abscisse+1)
        y.append(NIVEAU[0])
        abscisse += pas
        i=i+1
    
    while 2*m<=i and i<=t:
        #instant_t=rd.randint(0,t)
        #d=numpy.exp(-((instant_t**2)/2)/(2*3.1415)**(1/2))
        r=rd.randint(0,100000) #???
        p=rd.randint(900,1800)
        intensité=rd.randint(1,100)
        intensité=intensité/10000000
        
        
        if int(sécheresse)==1:
            intensité=intensité*20
            r=rd.randint(0,100000)
        
        if int(barrage)==1:
            intensité=intensité/2
            
        if 1<NIVEAU[0]<1.2:
            NIVEAU[0]=NIVEAU[0]*numpy.exp(-((a*D-((0.93*a)*D)))*(1/i)) #0.98=coefficient paramétrant la pente de l'exp décroissante
            x.append(abscisse+1)
            y.append(NIVEAU[0])
            abscisse += pas # on augmente abscisse de pas pour passer au point suivant
            i=i+1
            
        if r==1:
            for k in range(0,p):#définir p comme temps moyen d'une averse (ou pris au hasard?)
            
                NIVEAU[0]=NIVEAU[0]+NIVEAU[0]*intensité #trouver un bon coefficient (peut être progression arithmétique plus appropriée)
                x.append(abscisse+1)
                y.append(NIVEAU[0])
                abscisse += pas
                i=i+1
            
            
            
        
        
        
        if r!=1 and NIVEAU[0]>1.2:
            r=rd.randint(0,100000)
            NIVEAU[0]=NIVEAU[0]*numpy.exp(-((a*D)-D)*(1/i))
            x.append(abscisse+1)
            y.append(NIVEAU[0])
            abscisse += pas # on augmente abscisse de pas pour passer au point suivant
            i=i+1    
      
    plt.xlabel("Temps écoulé (en secondes)")
    plt.ylabel("Niveau d'eau du fleuve")
    plt.plot(x, y)
    plt.show()
    plt.close()
    
    return(NIVEAU[0], ETAT)
```
